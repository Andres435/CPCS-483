```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Import Dataset

```python
dataset = pd.read_csv('Data_regression.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

# Training the Simple Linear Regression model on Training set

```python
from sklearn. linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

# Calculate R2

```python
from sklearn.metrics import r2_score  # Import the r2_score function from scikit-learn

# Calculate the R-squared (R²) score
R2 = r2_score(y_test, y_pred)

# Print the R-squared value
print("R-squared (R²):", R2)

# You can calculate and print accuracy if you prefer to refer to it as such
accuracy = R2 * 100
print("Accuracy: {:.2f}%".format(accuracy))
```
